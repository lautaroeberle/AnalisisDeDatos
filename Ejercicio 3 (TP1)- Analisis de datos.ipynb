```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('total-usuarios-por-dia.csv')




colectivos_total = df['colectivo'].sum()
lanchas_total = df['lancha'].sum()
subtes_total = df['subte'].sum()
trenes_total = df['tren'].sum()

cantidades = [colectivos_total, lanchas_total, subtes_total, trenes_total]
etiquetas = ['Colectivos', 'Lanchas', 'Subtes', 'Trenes']


plt.figure(figsize=(8, 8))
plt.pie(cantidades, labels=etiquetas, autopct='%1.1f%%', startangle=140)
plt.title('Cantidad de usos de SUBE en transporte publico')
plt.axis('equal')
plt.show()



df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])


df['mes'] = df['indice_tiempo'].dt.month
lanchas_por_mes = df.groupby('mes')['lancha'].sum()


plt.figure(figsize=(10, 6))
lanchas_por_mes.plot(kind='bar', color='skyblue')
plt.title('Cantidad de usos de SUBE en lanchas por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de usos de SUBE')
plt.xticks(rotation=0)
plt.show()





transportes = ['Colectivos', 'Lanchas', 'Subtes', 'Trenes']
cantidades_2020 = [colectivos_2020, lanchas_2020, subtes_2020, trenes_2020]
cantidades_2023 = [colectivos_2023, lanchas_2023, subtes_2023, trenes_2023]

x = range(len(transportes))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x, cantidades_2020, width, label='2020', color='lightblue')
bars2 = ax.bar([p + width for p in x], cantidades_2023, width, label='2023', color='orange')

ax.set_xlabel('Tipo de Transporte')
ax.set_ylabel('Cantidad')
ax.set_title('Comparaci√≥n de Transportes entre 2020 y 2023')
ax.set_xticks([p + 0.5 * width for p in x])
ax.set_xticklabels(transportes)
ax.legend()


def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bars1)
autolabel(bars2)

plt.show()
```

El csv que eleg√≠ contiene los la cantidad de usos de SUBE por dia en diferentes transportes p√∫blicos desde el primer dia de 2020 hasta el 16/04/24.

El primer grafico muestra la cantidad de usos totales de SUBE en todos los transportes p√∫blicos. Se nota que el colectivo es sumamente mas usado que los demas transportes. Probablemente porque existen muchas mas unidades que recorren mas zonas del pais

El segundo grafico muestra la cantidad de usos de SUBE en lanchas durante los diferentes meses del a√±o. La conclusion a la que llego es que el uso de sube en lanchas se reduce en mayo, junio, julio y agosto porque esos meses son frios y la gente prefiere no navegar con ese clima

El tercer grafico compara la cantidad de usos de SUBE en 2020 con la cantidad de usos de SUBE en 2023. En el grafico se que como los usos de SUBE en 2023 superan ampliamente a los de 2020. Llego a la conclusion de que esto es asi debido a que en 2020, plena pandemia, mucha gente hizo home office y otros muchos directamente no podian ir a trabajar, por lo que no tenian la necesidad de desplazarse de sus casas
