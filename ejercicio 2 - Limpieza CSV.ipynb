```python
import pandas as pd
import numpy as np
df = pd.read_csv('occupation-2018-census-csv.csv')
# la longitud del codigo debe ser de 6 caracteres
df['Code'] = df['Code'].astype(str).str.zfill(6)
# pongo todos los valores de la columna de ocupaciones en minuscula
df['Occupation'] = df['Occupation'].str.lower().str.strip()
#reemplazo valores nulos por 0
df['Employed_census_usually_resident_population_count_aged_15_years_and_over'].fillna(0, inplace=True)
# elimino filas que contengan valores nulos en cualquier columna
df.dropna(inplace=True)

#nuevo csv
cleaned_file_path = 'censoDeOcupacionLimpio.csv'
df.to_csv(cleaned_file_path, index=False)


```
