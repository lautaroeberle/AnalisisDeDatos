```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./csvlimpio.csv')

prioridades = {
    'Tamaño de las porciones': 0,
    'Calidad de la comida': 0,
    'Precios': 0,
    'Reputacion del lugar': 0
}

def obtenerDatosPrioridades(valor):
  global prioridades
  i = 0
  prioridadActual = ""
  while(i < len(valor)):
    while(i < len(valor) and valor[i] != ';'):
      prioridadActual += valor[i]
      i += 1
    prioridades[prioridadActual] += 1
    prioridadActual = ''
    i += 1

df['Prioridades'].apply(obtenerDatosPrioridades)

plt.bar(['Tamaño de las porciones', 'Calidad de la comida', 'Precios', 'Reputación del lugar'], prioridades.values())
plt.xticks(rotation = 45)
plt.xlabel('Cuestiones priorizadas')
plt.ylabel('Cantidad de gente')
plt.title('Cuestiones priorizadas de un lugar')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./csvlimpio.csv')

afecciones = {
    'Dolor de panza': 0,
    'Descompostura': 0,
    'Arcadas/vomitos': 0,
    'Fiebre': 0,
    'Atracón': 0,
    'Malestar estomacal': 0,
    'Malestar general': 0,
    'Ninguna de las mencionadas': 0
}

def obtenerDatosAfecciones(valor):
  global afecciones
  i = 0
  afeccionActual = ""
  while(i < len(valor)):
    while(i < len(valor) and valor[i] != ';'):
      afeccionActual += valor[i]
      i += 1
    if afeccionActual in afecciones:
      afecciones[afeccionActual] += 1
    afeccionActual = ''
    i += 1

df['Afecciones posteriores'].apply(obtenerDatosAfecciones)

plt.bar(['Dolor de panza', 'Descompostura', 'Arcadas / vómitos', 'Fiebre', 'Atracón', 'Malestar estomacal', 'Malestar general', 'Ninguna'], afecciones.values())
plt.xticks(rotation = 45)
plt.xlabel('Afecciones')
plt.ylabel('Cantidad de gente')
plt.title('Afecciones posteriores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./csvlimpio.csv')

respuestas = {
    'Si. Por esa razón dejé de ir al lugar donde comí / pedir lo que pedí.': 0,
    'Si, pero sigo yendo al lugar donde comí /pidiendo lo que pedí': 0,
    'No': 0,
    'No he padecido ninguna afección luego de comer afuera.': 0
}

def obtenerDatosRespuestas(valor):
  global respuestas
  i = 0
  respuestaActual = ""
  while(i < len(valor)):
    while(i < len(valor) and valor[i] != ';'):
      respuestaActual += valor[i]
      i += 1
    respuestas[respuestaActual] += 1
    respuestaActual = ''
    i += 1

df['¿Cree que la comida que había consumido afuera fue el causante de esas afecciones? '].apply(obtenerDatosRespuestas)

labels = ['Cree que fue esa comida,\ny dejó de ir al lugar', 'Cree que fue esa comida,\npero sigue yendo al lugar', 'No cree que haya\nsido esa comida', 'No sufrió afecciones']
sizes = respuestas.values()
colors = ['blue', 'orange', 'yellow', 'gray']
fig, ax = plt.subplots(figsize=(5, 5))

ax.pie(sizes, labels = labels, autopct = '%d%%', startangle = 90, colors = colors, pctdistance = 0.5)
plt.title('Opinión acerca de la causa de las afecciones')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./csvlimpio.csv')

cantidadGentePorGrupo = {
    'No': 0,
    'Si': 0
}

for i, valor in df['Enfermedad/condición'].items():
  if df['Lugares frecuentados'][i] == 'Cadenas de comida rápida' or df['Lugares frecuentados'][i] == 'Puestos en la calle':
    cantidadGentePorGrupo[valor] += 1;

plt.bar(['Sin enfermedades / condiciones', 'Con enfermedades / condiciones'], cantidadGentePorGrupo.values())
plt.xticks(rotation = 45)
plt.xlabel('Grupo de personas')
plt.ylabel('Cantidad de gente')
plt.title('Cantidad de gente con y sin enfermedades/condiciones\n que consume comida rápida y/o callejera')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./csvlimpio.csv')

afeccionesPorTipoAlimento = {
    'Alimentos hervidos (Pastas, vegetales hervidos, etc.)': 0,
    'Alimentos horneados o hechos a la plancha(Carnes, vegetales horneados, etc.)': 0,
    'Alimentos crudos (Vegetales crudos, enlatados, etc.)': 0,
    'Alimentos fritos (Papas fritas, empanados fritos, etc)': 0
}

for i, valor in df['Afecciones posteriores'].items():
  if valor != 'Ninguna de las mencionadas':
    tiposAlimento = df['Tipo de alimentos'][i]
    tipoActual = ''
    j = 0
    while(j < len(tiposAlimento)):
      while(j < len(tiposAlimento) and tiposAlimento[j] != ';'):
        tipoActual += tiposAlimento[j]
        j += 1
      afeccionesPorTipoAlimento[tipoActual] += 1
      tipoActual = ''
      j += 1

plt.bar(['Hervidos', 'Horneados / a la plancha', 'Crudos', 'Fritos'], afeccionesPorTipoAlimento.values())
plt.xticks(rotation = 45)
plt.xlabel('Tipo de alimentos')
plt.ylabel('Cantidad de afecciones')
plt.title('Cantidad de afecciones por tipo de comida consumida')
plt.show()
```
